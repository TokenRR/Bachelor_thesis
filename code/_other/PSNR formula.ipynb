Отримання значень PSNR: `inf` та MSE: `0.0` свідчить про те, що між оригінальним та покращеним зображенням немає відмінностей (вони ідентичні).

Щоб зрозуміти, чому PSNR відображається як `inf`, слід розглянути формулу PSNR:

$$
PSNR = 10 \cdot \log_{10}\left(\frac{{\text{{data range}}^2}}{{\text{{MSE}}}}\right)
$$

У випадку, коли MSE дорівнює 0 (або дуже близьке до 0), в знаменнику у формулі виходить дуже маленьке число, тенденція до нуля. У такому випадку \( $\log_{10}(0)$ \) є нескінченності, тобто PSNR стає нескінченно великим або "inf". Це вказує на те, що між оригінальним та покращеним зображенням немає помилок (MSE = 0).

Якщо ви хочете отримати PSNR у вигляді числа, можливо, вам варто перевірити, чому ваші оригінальне та покращене зображення виглядають ідентично. Можливо, виникає яка-то помилка у процесі зчитування чи обробки зображень. Переконайтеся, що ви використовуєте різні файли зображень для порівняння.
